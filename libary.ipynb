```python
class Book:
    def __init__(self, book_name, author_name, price, available_copies):
        self.book_name = book_name
        self.author_name = author_name
        self.price = price
        self.available_copies = available_copies 
        

    def borrow_book(self):
        if self.available_copies > 0:
            self.available_copies -= 1
            print("Book Borrowed Succesflly")
        else:
            print("Book not available")
    def return_book(self):
        self.available_copies += 1
        print("Book Return Succesflly")

    def display(self):
        print("Book Name:", self.book_name)
        print("Author Name:", self.author_name)
        print("Price:", self.price)
        print("Available Copies:", self.available_copies)
```

```python
book1 = Book(
    "Pyhton",
    "Saad",
    2000,
    5
)

book1.display()
```

```python
print("\nBorrowing Book...\n")
book1.borrow_book()
book1.display()
```

```python
print("\nReturning Book...\n")
book1.return_book()
book1.display()
```
