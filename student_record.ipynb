```python
class Student:
    def __init__(self,name,roll_no,computer_marks,math_marks,english_marks):
        self.name = name
        self.roll_no = roll_no
        self.computer_marks = computer_marks
        self.math_marks = math_marks
        self.english_marks = english_marks

    def calculate_total(self):
        return self.computer_marks + self.math_marks + self.english_marks
    
    def calculate_percentage(self):
        total = self.calculate_total()
        return (total/300)*100
    
    def calculate_grade(self):
        percentage = self.calculate_percentage()

        if percentage >= 80:
            return "A"
        elif percentage >= 70:
            return "B"
        else:
           return "Fail"

    def show(self):
        print("name : ", self.name)        
        print("rollno : ", self.roll_no)        
        print("computer marks : ", self.computer_marks)        
        print("math marks : ", self.math_marks)        
        print("english marks : ", self.english_marks)        
        print("Toatal Marks : ", self.calculate_total())        
        print("Percentage : ", self.calculate_percentage())        
        print("Grade : ", self.calculate_grade())        
```

```python
student_name = input("enter student name: ")
student_rollno = input("enter student roll no: ")
computer_marks = int(input("enter computer marks: "))
math_marks = int(input("enter math marks: "))
english_marks = int(input("enter english marks: "))

studen_record = Student(student_name, student_rollno,computer_marks,math_marks,english_marks) 
studen_record.show()

```
