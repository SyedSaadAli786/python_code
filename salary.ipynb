```python
class Employe:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def salary_employe(self):
        return self.salary
    def display(self):
        print("Employe Name: " , self.name)
        print("Salary " , self.salary_employe())

class Manager(Employe):
    def __init__(self, name, salary,bouns):
        super().__init__(name, salary)
        self.bouns = bouns
    def salary_employe(self):
        return super().salary_employe() + self.bouns    
class Developer(Employe):
    def __init__(self, name, salary,bouns):
        super().__init__(name, salary)
        self.bouns = bouns
    def salary_employe(self):
        return super().salary_employe() + self.bouns    

        
```

```python
manger1 = Manager("Saad", 80000,20000)
developer1 = Developer("Ali", 60000,10000)

print("----- Manager -----")
manger1.display()
print("----- Developer -----")
developer1.display()

```
