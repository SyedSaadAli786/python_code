```python
class Employee:

    def __init__(self, name, basic_salary):
        self.name = name
        self.basic_salary = basic_salary

    def calculate_salary(self):
        return self.basic_salary

    def display(self):
        print("Employee Name:", self.name)
        print("Salary:", self.calculate_salary())


class Manager(Employee):

    def __init__(self, name, basic_salary, bonus):
        super().__init__(name, basic_salary)
        self.bonus = bonus

    # def calculate_salary(self):
    #     return self.basic_salary + self.bonus


class Developer(Employee):

    def __init__(self, name, basic_salary, overtime):
        super().__init__(name, basic_salary)
        self.overtime = overtime

    def calculate_salary(self):
        return self.basic_salary + self.overtime


# Objects

manager1 = Manager("Saad", 80000, 20000)
developer1 = Developer("Ali", 70000, 10000)

print("----- Manager -----")
manager1.display()

print()

print("----- Developer -----")
developer1.display()
```
